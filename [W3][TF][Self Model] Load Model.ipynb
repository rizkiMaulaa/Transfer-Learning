```python
#library untuk load model
from keras.models import load_model
```

```python
#menginisiasi var model_baru dengan hasil load modelku.h5
model_baru = load_model('modelku.h5')
```

```python
#library untuk proses test
import requests
from io import BytesIO

from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Parameters
input_size = (50,50)

#define input shape
channel = (3,)
input_shape = input_size + channel

#define labels
labels = ['paper', 'rock', 'scissor']
```

```python
def preprocess(img,input_size):
    nimg = img.convert('RGB').resize(input_size, resample= 0)
    img_arr = (np.array(nimg))/255
    return img_arr

def reshape(imgs_arr):
    return np.stack(imgs_arr, axis=0)
```

```python
im = Image.open("/content/drive/My Drive/DATA ML/[W1]RPS/datatest/Zx5NMbS4YBf29guF.png")
X = preprocess(im,input_size)
X = reshape([X])
y = model_baru.predict(X)
plt.imshow(im)
print( labels[np.argmax(y)], np.max(y) )
```
